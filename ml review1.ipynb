```python
columns_to_modify = [f'attr{i}' for i in range(1, 295) if f'attr{i}' in df.columns]

# Add missing values to the first 10 rows in specified columns only
num_missing_per_row = 1  # Number of missing values to add per row

for row in range(10):  # Iterate over the first 10 rows
    # Randomly select columns to insert missing values from the specified range
    cols_to_nan = np.random.choice(columns_to_modify, num_missing_per_row, replace=False)
    for col in cols_to_nan:
        df.at[row, col] = np.nan

# Display the first 10 rows of the dataset to verify missing values
print("First 10 rows with missing values:")
print(df.head(10))
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)

# Total number of missing values
total_missing = missing_values.sum()
print("\nTotal number of missing values:", total_missing)
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)

# Fill missing values in object (categorical) columns with the mode
for column in df.select_dtypes(include=['object']).columns:
    df[column].fillna(df[column].mode()[0], inplace=True)

# Verify that there are no more missing values
missing_values_after = df.isnull().sum()
print("\nMissing values after imputation:")
print(missing_values_after)

# Display the first 10 rows to verify
print(df.head(10))
```

```python
# Identify categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
print(categorical_columns)
```

```python
# One-hot encode categorical variables
# Encoding categorical variables
df_encoded = pd.get_dummies(df, columns=categorical_columns)
print(df_encoded.head())
```

```python
#splitting the data for training and testing dataset
from sklearn.model_selection import train_test_split

# Split the data into features and labels
features = df_encoded.drop(columns=['Beach_b\'0\'', 'Beach_b\'1\'', 'Sunset_b\'0\'', 'Sunset_b\'1\'', 'FallFoliage_b\'0\'', 'FallFoliage_b\'1\'', 'Field_b\'0\'', 'Field_b\'1\'', 'Mountain_b\'0\'', 'Mountain_b\'1\'', 'Urban_b\'0\'', 'Urban_b\'1\''])
labels = df_encoded[['Beach_b\'0\'', 'Beach_b\'1\'', 'Sunset_b\'0\'', 'Sunset_b\'1\'', 'FallFoliage_b\'0\'', 'FallFoliage_b\'1\'', 'Field_b\'0\'', 'Field_b\'1\'', 'Mountain_b\'0\'', 'Mountain_b\'1\'', 'Urban_b\'0\'', 'Urban_b\'1\'']]

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Display the shapes of the training and testing sets
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Import necessary libraries for modeling
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Train a Decision Tree model with shallow depth (to cause underfitting)
underfit_model = DecisionTreeClassifier(max_depth=1, random_state=42)
underfit_model.fit(X_train, y_train)

# Make predictions on the training and testing sets
y_train_pred_underfit = underfit_model.predict(X_train)
y_test_pred_underfit = underfit_model.predict(X_test) 

# Calculate accuracy on the training and testing sets
train_accuracy_underfit = accuracy_score(y_train, y_train_pred_underfit)
test_accuracy_underfit = accuracy_score(y_test, y_test_pred_underfit)

print(f"Underfitting Model - Training Accuracy: {train_accuracy_underfit}")
print(f"Underfitting Model - Testing Accuracy: {test_accuracy_underfit}")
```

```python
# Train a Decision Tree model with high depth (to cause overfitting)
overfit_model = DecisionTreeClassifier(max_depth=None, random_state=42)
overfit_model.fit(X_train, y_train)

# Make predictions on the training and testing sets
y_train_pred_overfit = overfit_model.predict(X_train)
y_test_pred_overfit = overfit_model.predict(X_test)

# Calculate accuracy on the training and testing sets
train_accuracy_overfit = accuracy_score(y_train, y_train_pred_overfit)
test_accuracy_overfit = accuracy_score(y_test, y_test_pred_overfit)

print(f"Overfitting Model - Training Accuracy: {train_accuracy_overfit}")
print(f"Overfitting Model - Testing Accuracy: {test_accuracy_overfit}")
```

```python
# Fit a linear regression model
linear_model = LinearRegression()
linear_model.fit(X, y)
```

```python
# Extract the coefficients (m) and the intercept (c)
coefficients = linear_model.coef_
intercept = linear_model.intercept_

# Display the equation y = mx + c for each feature
print("Linear Regression Equation:")
for feature, coef in zip(X.columns, coefficients):
    print(f"{feature}: y = {coef} * x + {intercept}")
```

```python
# Select one independent variable (e.g., the first one in the dataset)
selected_feature = X.columns[0]
x = X[selected_feature]
y = df_encoded['Urban_b\'1\'']

# Calculate the mean of x and y
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate the slope (m) using the given formula
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)
m = numerator / denominator

# Calculate the intercept (c) using the formula c = y_mean - m * x_mean
c = y_mean - m * x_mean

print(f"Slope (m): {m}")
print(f"Intercept (c): {c}")

# Equation of the line
y_pred = m * x + c

# Plot the data points and the regression line
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, y_pred, color='red', label=f'Line: y = {m:.2f}x + {c:.2f}')
plt.xlabel(selected_feature)
plt.ylabel('Urban_b\'1\'')
plt.title(f'Linear Relationship between {selected_feature} and Urban_b\'1\'')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Initialize the KFold and the model
kf = KFold(n_splits=5, shuffle=True, random_state=42)
linear_model = LinearRegression()

# Create a figure to plot the results for each fold
plt.figure(figsize=(15, 10))

# Perform K-Fold Cross-Validation
fold = 1
for train_index, test_index in kf.split(X):
    # Split the data into training and testing sets for this fold
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Train the model on the training set
    linear_model.fit(X_train, y_train)
    
    # Predict on the testing set
    y_pred = linear_model.predict(X_test)
    
    # Plot actual vs. predicted values
    plt.subplot(3, 2, fold)
    plt.scatter(y_test, y_pred, alpha=0.7)
    plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2)
    plt.xlabel('Actual')
    plt.ylabel('Predicted')
    plt.title(f'Fold {fold}: Actual vs Predicted')
    plt.grid(True)
    
    fold += 1

plt.tight_layout()
plt.show()
```
